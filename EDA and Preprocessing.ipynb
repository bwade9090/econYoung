## Original Panel

```python
years = [i for i in range(1, 14)]
```

```python
ids = [i for i in range(1, 10207)]
```

```python
multi_index = pd.MultiIndex.from_product([ids, years], names=["id", "year"])
```

```python
multi_index
```

```python
basic_var_names = list(pd.read_csv('data/통합변수명_수정.csv')['name'])
basic_var_names
```

```python
basic_var_names.pop(0)
basic_var_names
```

```python
individual_var_names = list(pd.read_csv('data/개별변수명.csv')['name'])
individual_var_names
```

```python
var_names = basic_var_names + individual_var_names
```

```python
original_panel_data = pd.DataFrame(index=multi_index, columns=var_names)
```

```python
original_panel_data
```

```python
df_dict = {}
for y in years:
  df_dict[y] = pd.read_csv('data/raw/ypdata_w{0:02d}.csv'.format(y), index_col='sampid')
```

```python
for id in ids:
  for year in years:
    original_panel_data.loc[(id, year)]['w**'] = df_dict[13].loc[id][f'w{year}']
    original_panel_data.loc[(id, year)]['gender'] = df_dict[13].loc[id]['gender']
    original_panel_data.loc[(id, year)]['yob'] = df_dict[13].loc[id][f'yob']
```

```python
import pickle
```

```python
with open('data/original_panel_data.pkl', 'wb') as f:
  pickle.dump(original_panel_data, f)
```
