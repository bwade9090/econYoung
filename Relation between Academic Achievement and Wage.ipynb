```python
import statsmodels.api as sm
from statsmodels.formula.api import ols
from bioinfokit.analys import stat
import matplotlib.pyplot as plt
import scipy.stats as stats
```

### 원데이터

```python
# interaction 무시
formula_str = 'y13wage ~ C(y02a060) + C(y02a061) + C(y02a064)'
res = stat()
res.anova_stat(df=df.astype({'y02a060':'category','y02a061':'category','y02a064':'category'}), res_var='y13wage', anova_model=formula_str)
res.anova_summary
```

```python
sm.qqplot(res.anova_std_residuals, line='45')
plt.xlabel("Theoretical Quantiles")
plt.ylabel("Standardized Residuals")
plt.show()
```

```python
# histogram
plt.hist(res.anova_model_out.resid, bins='auto', histtype='bar', ec='k') 
plt.xlabel("Residuals")
plt.ylabel('Frequency')
plt.show()
```

```python
w, pvalue = stats.shapiro(res.anova_model_out.resid)
print(w, pvalue)
```

```python
res = stat()
res.levene(df=df.astype({'y02a060':'category','y02a061':'category','y02a064':'category'})[['y13wage', 'y02a060', 'y02a061', 'y02a064']], res_var='y13wage', xfac_var=['y02a060','y02a061','y02a064'])
res.levene_summary
```

```python
res = stat()
res.levene(df=df.astype({'y02a060':'category','y02a061':'category','y02a064':'category'})[['y13wage', 'y02a060', 'y02a061', 'y02a064']], res_var='y13wage', xfac_var='y02a060')
res.levene_summary
```

### 수학 수준 재조정

```python
def convert_to_quantile(level):
  if level == 1:
    return 0.05
  elif level == 2:
    return 0.2
  elif level == 3:
    return 0.5
  elif level == 4:
    return 0.8
  elif level == 5:
    return 0.95
  else:
    return np.nan
```

```python
df_no_outlier_cleaned = df_no_outlier_cleaned.copy()
df_no_outlier_cleaned['math_quentile'] = df_no_outlier_cleaned.y02a061.apply(convert_to_quantile)
```

```python
sns.lineplot(data=df_no_outlier_cleaned, x="math_quentile", y="y13logwage")
```
