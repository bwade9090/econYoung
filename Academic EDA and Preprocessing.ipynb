## 2019년(13차)

```python
# 임금유형
sns.countplot(df_13.y13b224)
```

```python
df_13['y13wage'] = np.nan
```

```python
def convert_to_annual_salary(row):
  if row['y13b224'] == 1:
    row['y13wage'] = row['y13b225']
  elif row['y13b224'] == 2:
    row['y13wage'] = row['y13b225'] * 12
  elif row['y13b224'] == 3:
    row['y13wage'] = row['y13b225'] * 52
  elif row['y13b224'] == 4:
    row['y13wage'] = row['y13b225'] * 52 * 5
  elif row['y13b224'] == 5:
    row['y13wage'] = row['y13b225'] * 52 * 40
  else:
    row['y13wage'] = np.nan
  
  if row['y13b225'] > 9e6:
    row['y13wage'] = np.nan
  
  return row
```

```python
df_13 = df_13.apply(convert_to_annual_salary, axis=1)
```

```python
df_13.y13wage.max()
```

```python
df_13.y13wage.min()
```

```python
# 임금
sns.boxplot(df_13.y13wage)
```

```python
# 학업 관련 columns 제거
df_13 = df_13.drop(df_13.filter(regex='^y.*a[0-9]', axis=1).columns, axis=1)
```
